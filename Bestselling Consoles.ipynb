# Consoles by Company 

```python
fig = plt.figure(figsize=(8,8), facecolor="#fbfbfb")
gs = fig.add_gridspec(1,1)
gs.update

sns.countplot(df["Company"], color="#14F3F0")

plt.xlabel("")
plt.xticks(rotation=90)
plt.ylabel("")

plt.grid(axis="y", linestyle="-.", alpha=0.7)

plt.text(-0.25, 18, "Consoles by Company", fontsize=20 )
```

# Sales by Company

```python
sales_by_company = df.groupby("Company")["Units sold (million)"].sum().reset_index().sort_values(
    by="Units sold (million)", ascending=False)
sales_by_company
```

```python
fig = plt.figure(figsize=(8,8), facecolor="#fbfbfb")
gs = fig.add_gridspec(1,1)
gs.update

sns.barplot(x="Units sold (million)", y="Company", data=sales_by_company, color="#14F3F0", orient="h")

plt.xlabel("")
plt.ylabel("")

plt.grid(axis="x", linestyle="-.", alpha=0.7)

plt.text(0, -0.55, " Units Sold by Company", fontsize=20)
```

# Releases by Year vs Discontinuation

```python
Games_by_Year = df[["Released Year", "Discontinuation Year"]]
Games_by_Year
```

```python
fig = plt.figure(figsize=(12,8), facecolor="#fbfbfb")
gs = fig.add_gridspec(1,1)
gs.update()

ax = fig.add_subplot(gs[0,0])

colors = {"Released Year": "#0D97A3", "Discontinuation Year": "#F31417"}

Games_by_Year.plot(ax=ax, color=colors)

plt.grid(True, linestyle="-.")

plt.title("Released Year vs Discontinuation Year", loc="left", fontsize=20)
```

# Games being released and discontinued by year 

```python
released = df["Released Year"].value_counts().reset_index().head(10)
discontinued = df["Discontinuation Year"].value_counts().reset_index().head(10)

print(released)
print(discontinued)
```

```python
fig = plt.figure(figsize=(12,10), facecolor="#fbfbfb")
gs = fig.add_gridspec(1,2)
gs.update()

ax0 = fig.add_subplot(gs[0,0])
ax1 = fig.add_subplot(gs[0,1])

sns.barplot(x="index", y="Released Year", data=released, color="#14F3F0" , ax=ax0)
sns.barplot(x="index", y="Discontinuation Year", data=discontinued, color="#F31417", ax=ax1)

ax0.set_xlabel("")
ax1.set_xlabel("")

plt.suptitle("Released Year vs Discontinued Year", fontsize=20, fontweight="bold")
ax0.text(-0.5, 4.25, "Top 10 Years by released Consoles", fontsize=15)
ax1.text(-0.5, 4.25, "Top 10 Year by discontinued Consoles", fontsize=15)
```

# Ongoing Consoles

# Comparing the consoles

```python
# sales comparison
fig = plt.figure(figsize=(16,8), facecolor="#fbfbfb")
gs = fig.add_gridspec(1,3)
gs.update()

colors = {"Nintendo Switch": "#7AAEE9", "PlayStation 4": "#232629", "PlayStation 5": "#E6EAED",
          "Xbox Series X/S" : "#45F277" }

ax0 = fig.add_subplot(gs[0,0])
ax1 = fig.add_subplot(gs[0,1])

sns.barplot(x="Console Name", y="Units sold (million)", data=on_going_consoles,palette=colors, ax=ax0)
sns.barplot(x="Console Name", y="Released Year", data=on_going_consoles, palette=colors, ax=ax1)

ax0.set_xlabel("")
ax0.set_ylabel("* million")
ax1.set_xlabel("")
ax1.set_ylabel("")

ax0.tick_params(labelrotation=90)
ax1.tick_params(labelrotation=90)

ax1.set_ylim(2010, 2023)

plt.suptitle("On going consoles Camparison", fontsize=20, fontweight="bold", ha="right")
ax0.text(-0.5, 129.5, "Sales by ongoing Consoles", fontsize=15)
ax1.text(-0.5,2023.05, "Release Year of ongoing Consoles",fontsize=15)
```

# Go into more detail of a company by your choice 

```python
# choose any company by your choice 
nintendo = df[df["Company"] == "Nintendo"]
nintendo
```

```python
# Sold units by console
fig = plt.figure(figsize=(12,8), facecolor="#fbfbfb")
gs = fig.add_gridspec(1,1)
gs.update()

sns.barplot(x="Units sold (million)", y="Console Name", data=nintendo.sort_values(by="Units sold (million)",
            ascending=False), color="#D71D2D")

plt.grid(axis="x", linestyle="-.", alpha=0.7)

plt.ylabel("")

plt.title("Nintendo Consoles by Units Sold", loc="left", va="top", fontsize=20)
```

```python
# Console Type 
fig = plt.figure(figsize=(12,8), facecolor="#fbfbfb")
gs = fig.add_gridspec(1,1)
gs.update()

sns.countplot(nintendo["Type"], color="#D71D2D")

plt.grid(axis="y", linestyle="-.", alpha=0.7)

plt.xlabel("")
plt.ylabel("")

plt.title("Console Type", loc="left", va="top", fontsize=20)
```

```python
# release year vs discontinuation Year
fig, ax1 = plt.subplots(figsize=(12, 8), facecolor="#fbfbfb")
ax2 = ax1.twinx()

nintendo1 = nintendo.copy()
nintendo1 = nintendo1.sort_values(by="Released Year")
nintendo2 = nintendo1.sort_values(by="Discontinuation Year")

ax1.plot(nintendo1["Released Year"], color='#D71D2D', label='Released Year')
ax2.plot(nintendo2["Discontinuation Year"], color='#000000', label='Discontinuation Year')

ax1.set_xlabel('Year')
ax1.set_ylabel('Released Year', color='#D71D2D')
ax2.set_ylabel('Discontinuation Year', color='#000000')

ax1.set_title('Nintendo Releases and Discontinuations Over Time', fontsize=20)

ax1.grid(True, linestyle="-.")

ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.show()
```

```python
# count by consoles realeses and discontinuations
fig, ax1 = plt.subplots(figsize=(12, 8), facecolor="#fbfbfb")
ax2 = ax1.twinx()

nintendo1 = nintendo.copy()

nintendo1["Released Year"].value_counts().sort_index().plot(ax=ax1, color='#D71D2D', label='Released Year')
nintendo1["Discontinuation Year"].value_counts().sort_index().plot(ax=ax2, color='#000000', label='Discontinuation Year')

ax1.set_xlabel('Year')
ax1.set_ylabel('Count of Released Consoles', color='#D71D2D')
ax2.set_ylabel('Count of Discontinued Consoles', color='#000000')

ax1.set_title('Nintendo Releases and Discontinuations Count by Year', fontsize=20)

ax1.grid(True, linestyle="-.")

ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 8), facecolor="#fbfbfb")

sns.lineplot(x="Released Year", y="Units sold (million)", data=nintendo,
             color="#D71D2D")

ax.grid(axis="both", linestyle="-.", alpha=0.7)
ax.set_xlabel("Year")
ax.set_ylabel("Units sold (million)")
ax.set_title("Nintendo Consoles by Units Sold Over Time", loc="left", va="top", fontsize=20)

plt.show()
```
