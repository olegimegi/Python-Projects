```python
# changing column names for easier reading
majors.columns = ["Major", "start_p50", "mid_p50", "increase", "mid_p10", "mid_p25", "mid_p75", "mid_p90"]
majors.head()
```

```python
# converting columns to float 
dollar_cols = ['start_p50','mid_p50','mid_p10','mid_p25','mid_p75','mid_p90']

for i in dollar_cols:
    majors[i] = majors[i].str.replace("$", "")
    majors[i] = majors[i].str.replace(",", "")
    majors[i] = pd.to_numeric(majors[i])

majors.head()
```

# Visualizing the data 

```python
# visualizing the average starting salary by major
fig = plt.figure(figsize=(5,10))

x = majors["start_p50"]
y = majors["Major"]

plt.scatter(x, y, label="Average Starting Salary by Majors")

plt.xlabel("US $")
plt.ylabel("")
plt.title("Starting Salary with Python")
plt.legend(loc=2) 
plt.show()
```
